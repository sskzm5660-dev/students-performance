```python
import pandas as pd
df=pd.read_csv("D:\student\students-performance\StudentsPerformance.csv")
```

```python
df
```

```python
df["Average Score"]=(df["math score"]+df["reading score"]+df["writing score"])/3
```

```python
df
```

```python
df.drop(columns=["math score","reading score","writing score"],inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gender_avg = df.groupby('gender')['Average Score'].mean()
```

```python
gender_avg
```

```python
gender_avg.plot(kind='bar')

plt.title("Gender vs Average Score")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
sns.histplot(data=df, x='Average Score', hue='gender', kde=True)

plt.title("Histogram of Average Score by Gender")
plt.xlabel("Average Score")
plt.ylabel("Count")
plt.show()


```

```python
race_avg = df.groupby('race/ethnicity')['Average Score'].mean()
```

```python
race_avg.plot(kind='bar')

plt.title("Race vs Average Score")
plt.xlabel("Race")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.pie(race_avg, labels=race_avg.index, autopct='%1.1f%%')
plt.title("Distribution of Average Scores by Race/Ethnicity")
plt.show()
```

```python
education= df.groupby('parental level of education')['Average Score'].mean()
```

```python
education.plot(kind='bar')
```

```python
plt.pie(education, labels=education.index, autopct='%1.1f%%')
```
